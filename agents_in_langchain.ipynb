```python

from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
import requests
```

```python
from langchain_community.tools import DuckDuckGoSearchRun

search_tool = DuckDuckGoSearchRun()
```

```python
@tool
def get_weather_data(city: str) -> str:
  """
  This function fetches the current weather data for a given city
  """
  url = f'https://api.weatherstack.com/current?access_key=4d1d8ae207a8c845a52df8a67bf3623e&query={city}'

  response = requests.get(url)

  return response.json()
  
```

```python
from langchain_ollama import OllamaLLM
# llm = OllamaLLM(model="ministral-3:8b-cloud")
llm = ChatOpenAI()
```

```python
from langchain_classic.agents import create_react_agent, AgentExecutor # create react agent is a design pattern agent
from langchain_classic import hub # predefined prompts 
```

```python
# Step 2: Pull the ReAct prompt from LangChain Hub
prompt = hub.pull("hwchase17/react")  # pulls the standard ReAct agent prompt
```

```python
# Step 3: Create the ReAct agent manually with the pulled prompt
agent = create_react_agent(
    llm=llm,
    tools=[search_tool, get_weather_data],
    prompt=prompt
)
```

```python
# Step 4: Wrap it with AgentExecutor
agent_executor = AgentExecutor(
    agent=agent,
    tools=[search_tool, get_weather_data],
    verbose=True
)
```

```python
# Step 5: Invoke
response = agent_executor.invoke({"input": "who who played yesterdays(29 march 2026) ipl and what are the runs on each team"})
print(response)
```
